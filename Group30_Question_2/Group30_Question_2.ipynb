```python
import json
import pandas as pd


df_august = pd.read_json('../Group30_Question_2/august.json', 
                         lines=True)
df_july = pd.read_json('../Group30_Question_2/july.json', 
                         lines=True)
```

```python
df_july.shape
```

```python
df_july.head()
```

```python
df_july_new = df_july.drop('carpark_number', axis = 1)
df_july_new = df_july_new.fillna(0).astype('int64')
df_july_sum = df_july_new.sum(axis = 0)
df_july_sum  = df_july_sum.to_frame().reset_index()
df_july_sum.columns = ['TimeStamp', 'LotsAvailable']
df_july_sum[['hour', 'day']] = df_july_sum.TimeStamp.str.split("-",expand=True,)
df_july_sum = df_july_sum.drop('TimeStamp', axis = 1)
df_july_sum
```

Now let us import the test data and convert to the above format. 

```python
df_aug_new = df_august.drop('carpark_number', axis = 1)
df_aug_new = df_aug_new.fillna(0).astype('int64')
df_aug_sum = df_aug_new.sum(axis = 0)
df_aug_sum = df_aug_sum.to_frame().reset_index()
df_aug_sum.columns = ['TimeStamp', 'LotsAvailable']
df_aug_sum[['hour', 'day']] = df_aug_sum.TimeStamp.str.split("-",expand=True,)
df_aug_sum = df_aug_sum.drop('TimeStamp', axis = 1)
df_aug_sum
```

```python
X_train = df_july_sum.drop('LotsAvailable', axis = 1)
Y_train = df_july_sum['LotsAvailable'].to_frame()
X_test = df_aug_sum.drop('LotsAvailable', axis = 1)
Y_test = df_aug_sum['LotsAvailable'].to_frame()
```

```python
import numpy as np
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X_train, Y_train.values.ravel())
```

Let us now predict the data.

```python
y_pred = regressor.predict(X_test)
df_pred = pd.DataFrame(y_pred, columns=['Available Lots'])
df_pred['hour'] = df_aug_sum['hour']
df_pred['day'] = df_aug_sum['day']
df_pred
```
